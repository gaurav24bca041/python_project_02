```python
sns.countplot(data=df,x='Age Group',hue='Gender')
```

```python
ax = sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
gg = sns.countplot(data=df,x='Department',hue='Age Group')
sns.set(rc={'figure.figsize':(15,7)})
for bars in gg.containers:
    gg.bar_label(bars)
```

```python
d_state= df.groupby(['State'],as_index=False)['Department'].sum().sort_values(by='Department',ascending=False)
sns.set(rc={'figure.figsize':(15,15)})
sns.barplot(data=d_state,x='State',y='Department')
```

```python
ax = sns.countplot(data=df,x='State',hue='Department')
sns.set(rc={'figure.figsize':(20,7)})
for bars in ax.containers:
    ax.bar_label(bars)
```
